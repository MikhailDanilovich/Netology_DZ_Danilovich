Обязательная часть

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import time
```

```python
KEYWORDS = ['обзор', 'распознавание', 'поехали', 'wrike', 'join', 'железо']
```

```python
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
posts = soup.find_all('article', class_='post post_preview')
```

```python
needed_posts = pd.DataFrame()

for post in posts:
    for word in KEYWORDS:
        
        time.sleep(0.3)

        if word in post.find('a', class_='post__title_link').text.lower() or word in post.find('ul', class_='post__hubs').text.lower() or word in post.find('div', class_='post__text').text.lower():
            
            title_element = post.find('a', class_='post__title_link')
            
            date = post.find('span', class_='post__time').text
            head = title_element.text
            link = title_element.attrs.get('href')
        
            row = {'date': date, 'head': head, 'link': link}
            
            needed_posts = pd.concat([needed_posts, pd.DataFrame([row])])
            
            break

needed_posts            
```

Необязательная часть

```python
KEYWORDS = ['обзор', 'распознавание', 'поехали', 'wrike', 'join', 'железо']
```

```python
links = []
for post in posts: 
    links += [post.find('a', class_='post__title_link').attrs.get('href')]
```

```python
needed_posts = pd.DataFrame()

for link in links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.3)
    
    for word in KEYWORDS:
        if word in soup.find('span', class_='post__title-text').text.lower() or word in soup.find('ul', class_='post__hubs').text.lower() or word in soup.find('div', class_='post__body').text.lower():
            
            title_element = soup.find('span', class_='post__title-text')
            
            date = soup.find('span', class_='post__time').text
            head = title_element.text
        
            row = {'date': date, 'head': head, 'link': link}
            
            needed_posts = pd.concat([needed_posts, pd.DataFrame([row])])
            
            break

needed_posts
```

Задание 2 

Обязательная часть

```python
import json
```

```python
EMAIL = ['mdanilovich@yandex.ru', 'danilovich.echo@gmail.com', 'ya.chernov@gmail.com']
```

```python
URL = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
```

```python
base_of_breaches = pd.DataFrame()

headers = {
    'Vaar-Header-App-Product': 'hackcheck-web-avast',
    'Vaar-Version': '0'
}

for email in EMAIL:
    
    body = {
    'emailAddresses': [email]
    }
    
    breaches_for_email = requests.post(URL, headers=headers, data=json.dumps(body))

    time.sleep(0.3)
    
    for el in breaches_for_email.json()['breaches']: 
        date = breaches_for_email.json()['breaches'][el]['publishDate'][:10]
        source = breaches_for_email.json()['breaches'][el]['site']
        description = breaches_for_email.json()['breaches'][el]['description']
        
        row = {'email': email, 'date': date, 'source': source, 'description': description}
            
        base_of_breaches = pd.concat([base_of_breaches, pd.DataFrame([row])])
        
base_of_breaches    
```

Необязательная часть

```python
REQUEST = 'https://api.vk.com/method/wall.get?'
TOKEN = ''
```

```python
import time

import datetime
```

```python
params = {
    'access_token': TOKEN,
    'domain': 'netology',
    'count': 50,
    'v':'5.84'
}
```

```python
base_of_posts = pd.DataFrame()

res = requests.get(REQUEST, params)

for el in res.json()['response']['items']:
    
    time.sleep(0.3)

    date = datetime.datetime.fromtimestamp(el['date']).strftime('%Y-%m-%d %H:%M:%S')
    message = el['text']
    
    row = {'date': date, 'text': message}

    base_of_posts = pd.concat([base_of_posts, pd.DataFrame([row])])
    
base_of_posts
```
