```python
import pandas as pd
```

```python
horses = pd.DataFrame()

with open('horse_data.csv', 'r', encoding='utf-8') as f:
    for line in f:
        if line.split(',')[1] == '1':
            Age = 'Adult horse'
        elif line.split(',')[1] == '2':
            Age = 'Young (< 6 months)'
        else:
            Age = 'No information'
            
        if line.split(',')[12] == '1':
            distension = 'none'
        elif line.split(',')[12] == '2':
            distension = 'slight'
        elif line.split(',')[12] == '3':
            distension = 'moderate'
        elif line.split(',')[12] == '4':
            distension = 'severe'
        else:
            distension = 'No information'

        if line.split(',')[16] == '1':
            examination = 'normal'
        elif line.split(',')[16] == '2':
            examination = 'increased'
        elif line.split(',')[16] == '3':
            examination = 'decreased'
        elif line.split(',')[16] == '4':
            examination = 'absent'
        else:
            examination = 'No information'
            
        if line.split(',')[17] == '1':
            abdomen = 'normal'
        elif line.split(',')[17] == '2':
            abdomen = 'other'
        elif line.split(',')[17] == '3':
            abdomen = 'firm feces in the large intestine'
        elif line.split(',')[17] == '4':
            abdomen = 'distended small intestine'
        elif line.split(',')[17] == '5':
            abdomen = 'distended large intestine'
        else:
            abdomen = 'No information'
            
        if line.split(',')[20] == '1':
            appearance = 'clear'
        elif line.split(',')[20] == '2':
            appearance = 'cloudy'
        elif line.split(',')[20] == '3':
            appearance = 'serosanguinous'
        else:
            appearance = 'No information'
            
        if line.split(',')[3] == '?':
            temperature = None
        else:
            temperature = float(line.split(',')[3])
 
        if line.split(',')[21] == '?':
            protein = None
        else:
            protein = float(line.split(',')[21])
        
        row = {'Age': Age, 
               'Hospital Number': line.split(',')[2], 
               'rectal temperature': temperature, 
               'abdominal distension': distension, 
               'rectal examination': examination, 
               'abdomen': abdomen, 
               'abdominocentesis appearance': appearance, 
               'abdomcentesis total protein': protein}
        
        horses = pd.concat([horses, pd.DataFrame([row])])
        
# horses.head(15)
horses
```

Задание 2. Выбросы 


```python
q1 = horses['rectal temperature'].quantile(0.25)
q3 = horses['rectal temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horses[horses['rectal temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature')
remove_outliers
```

В выбросы попали лошади с температурой 36.7 и ниже и 39.55 и выше. При этом нормальная ректальная температура лошади - от 37,5 до 38,5. То есть в выбросы попали животные со значительным отклонением от нормы. Следовательно, их надо срочно лечить, если они еще живы. 

То есть исчисления помогли нам прикинуть, сколько у нас здоровых лошадей, а сколько мы пока не можем использовать.

```python
q1 = horses['abdomcentesis total protein'].quantile(0.25)
q3 = horses['abdomcentesis total protein'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horses[horses['abdomcentesis total protein'].between(lower_bound, upper_bound, inclusive=True)].sort_values('abdomcentesis total protein')
remove_outliers
```

Что касается показателя "abdomcentesis total protein", по нему есть данные только о 102 особях. Кажется, это очень мало, чтобы делать какие-то выводы.
